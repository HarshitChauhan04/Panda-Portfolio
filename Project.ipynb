```python
import numpy as np
import pandas as pd
```

```python
new=pd.DataFrame({'Product':['A','B','A','C','B','A','C','B','A','C'],
                  'Region':['North','South','North','East','East','South','North','South','East','North'],
                  'SalesPerson':['Alise','Bob','Charlie','Alise','Bob','Charlie','Alise','Bob','Charlie','Alise'],
                  'Units_Sold':[150,200,90,130,150,160,180,140,190,120],
                  'Revenue':[3000,4000,1800,2600,3000,3200,3600,2800,3800,2400]})
new
```

```python
x=pd.pivot_table(new,index='Product',columns='Region',values='Units_Sold',aggfunc='sum')
x
```

```python
x1=pd.pivot_table(new,index="Product",columns="SalesPerson",values="Revenue",aggfunc="mean")
x1
```

```python
x2=pd.pivot_table(new,index='SalesPerson',columns='Region',values='Units_Sold',aggfunc='max')
x2
```

```python
x3=pd.pivot_table(new,index='Region',values='Revenue',aggfunc=['sum','mean'])
x3
```

```python
x4=pd.pivot_table(new,index=['Product','Region'],values=['Revenue','Units_Sold'],aggfunc='sum')
x4
```

```python
x5=pd.pivot_table(new,index='SalesPerson',columns=['Product','Region'],values='Revenue',aggfunc='sum')
x5
```

```python
x6=pd.pivot_table(new,index='Region',values=['Revenue','Units_Sold'],aggfunc='sum')
x6
```

```python
df=pd.DataFrame({
    'Product':['Laptop','Laptop','TV','Phone','TV','Laptop','Phone','TV'],
    'Category':['Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics'],
    'Region':['North','South','North','South','West','West','North','South'],
    'Sales':[1000,1500,2000,1200,800,1300,1000,1700]
})
df
```

# Calculate the total sales per product for region north and south only

```python
df_temp=df[(df['Region']=='North')|(df['Region']=='South')]
# other method
df_temp=df[df['Region'].isin(['North','South'])]
df_temp           
```

```python
df1=pd.pivot_table(df_temp,index='Product',columns='Region',values='Sales',aggfunc='sum')
df1
```
