```python
import numpy as np
import pandas as pd
```

# add a new column in existing DataFrame

```python
x=pd.DataFrame({'Department':['Sales','Sales','IT','IT','Marketing','Marketing'],
                'Employee':['A','B','C','D','E','F'],
                'Salary':[1000,2000,5000,5000,3000,3000]})
x
                            
```

```python
year=pd.Series([2021,2022,2021,2022,2021,2022])
y=pd.DataFrame({'Year':year})
x=pd.concat([x,y],axis=1)
x
```

# Renaming a Column Name

```python
x.rename(columns={'Salary':'Monthly_Salary'},inplace=True)
x
```

# Pivot Table

```python
new=pd.DataFrame({'Date':['2024-11-01','2024-11-01','2024-11=02','2024-11=02','2024-11=03','2024-11=03'],
                  'Store':['A','B','A','B','A','B'],
                  'Product':['Pen','Pencil','Pen','Pencil','Pen','Pencil'],
                  'Sales':[100,200,300,400,500,600],
                  'Quantity':[10,20,30,40,50,60]})
new
```

```python
total=pd.pivot_table(new,index='Product',values='Sales',aggfunc=['sum','mean'])
total
```

```python
a=pd.pivot_table(new,index='Store',columns='Product',values='Sales',aggfunc='sum')
a
```
